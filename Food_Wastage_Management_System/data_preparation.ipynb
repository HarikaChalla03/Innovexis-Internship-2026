# **Data Cleaning**

```python
#Cleaning Date Datatypes
food["Expiry_Date"] = pd.to_datetime(food["Expiry_Date"], errors="coerce")
claims["Timestamp"] = pd.to_datetime(claims["Timestamp"], errors="coerce")
```

```python
food.info()
```

```python
claims.info()
```

```python
#Cleaning Providers Contact details column
import pandas as pd
import re

def clean_contact(contact):
    if pd.isna(contact):
        return None
    
    contact = str(contact)
    
    # Remove extension like x123
    contact = re.sub(r"x\d+", "", contact)
    
    # Keep only digits
    contact = re.sub(r"\D", "", contact)
    
    return contact  

# Apply cleaning
providers["Contact"] = providers["Contact"].apply(clean_contact)
```

```python
providers.head()
```

```python
#Cleaning receivers Contact details column
import pandas as pd
import re

def clean_contact(contact):
    if pd.isna(contact):
        return None
    
    contact = str(contact)
    
    # Remove extension like x123
    contact = re.sub(r"x\d+", "", contact)
    
    # Keep only digits
    contact = re.sub(r"\D", "", contact)
    
    return contact 

# Apply cleaning
receivers["Contact"] = receivers["Contact"].apply(clean_contact)
```

# **Expiry Based Features**

```python
import pandas as pd

providers = pd.read_csv("C:/Users/Harika challa/Innovexis/Project-2/providers_data.csv")
receivers = pd.read_csv("C:/Users/Harika challa/Innovexis/Project-2/receivers_data.csv")
food = pd.read_csv("C:/Users/Harika challa/Innovexis/Project-2/food_listings_data.csv")
claims = pd.read_csv("C:/Users/Harika challa/Innovexis/Project-2/claims_data.csv")
```

```python
food['Expiry_Date'] = pd.to_datetime(food['Expiry_Date'])

food["Days_To_Expire"] = (food["Expiry_Date"] - pd.Timestamp.today()).dt.days
food["Is_Expired"] = food["Days_To_Expire"] < 0
```

# **Claim Based Features**

```python
claims['Timestamp'] = pd.to_datetime(claims['Timestamp'])

claims['claim_hour'] = claims['Timestamp'].dt.hour
claims['claim_day'] = claims['Timestamp'].dt.day_name()
```

# **Merge Datasets**

```python
merged_datasets = food.merge(claims, on='Food_ID') \
                .merge(providers, on='Provider_ID') \
                .merge(receivers, on='Receiver_ID')
```

# **Save Merged Datasets**

```python
merged_datasets.to_csv("C:/Users/Harika challa/Innovexis/Project-2/final_merged_data.csv", index=False)
```
